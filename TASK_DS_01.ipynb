```python
#Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = 'Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_76034.csv'  
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
# Filter out rows with missing IncomeGroup values
income_data = df[df['IncomeGroup'].notna()]
income_data
```

```python
# Calculate frequency counts
income_counts = income_data['IncomeGroup'].value_counts().reset_index()
income_counts.columns = ['IncomeGroup', 'Count']
```

```python
# Plot using seaborn barplot
plt.figure(figsize=(8, 5))
sns.barplot(
    y='IncomeGroup',
    x='Count',
    data=income_counts,
    palette='mako'
)

# Add labels and title
plt.title('Distribution of Countries by Income Group')
plt.xlabel('Number of Countries')
plt.ylabel('Income Group')
plt.grid(True, axis='x')
plt.tight_layout()

# Show plot
plt.show()
```

```python
#OBSERVATIONS-
# High income group has the largest number of countries, with the bar extending beyond 80.
# Lower middle income and Upper middle income groups are close in number, each having around 50–55 countries.
# Low income group has the fewest countries, with fewer than 30.

#Insight-
# The chart shows that a higher number of countries fall into the high-income category, with progressively fewer in the middle
# and low-income groups.
```
